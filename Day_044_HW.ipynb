```python
from sklearn import datasets, metrics
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
```

```python
# 讀取資料集
boston = datasets.load_boston()

# 切分訓練集/測試集
x_train, x_test, y_train, y_test = train_test_split(boston.data, boston.target, test_size=0.25, random_state=4)

# 建立模型 - 隨機森林(使用 20 顆樹，每棵樹的最大深度為 4)
clf = RandomForestRegressor(n_estimators=20, max_depth=4)

# 訓練模型
clf.fit(x_train, y_train)

# 預測測試集
y_pred = clf.predict(x_test)
print(y_pred)
```

```python
# 分數 & MSE
print("The score of RandomForest : {}".format(clf.score(x_train, y_train)))
print("Mean squared error : {}".format(mean_squared_error(y_test, y_pred)))
```

```python
# 讀取資料集
boston = datasets.load_boston()

# 切分訓練集/測試集
x_train, x_test, y_train, y_test = train_test_split(boston.data, boston.target, test_size=0.25, random_state=4)

# 建立模型 - 線性回歸
clf = LinearRegression()

# 訓練模型
clf.fit(x_train, y_train)

# 預測測試集
y_pred = clf.predict(x_test)
print(y_pred)
```

```python
# 分數 & MSE
print("The score of Linear : {}".format(clf.score(x_train, y_train)))
print("Mean squared error : {}".format(mean_squared_error(y_test, y_pred)))
```
